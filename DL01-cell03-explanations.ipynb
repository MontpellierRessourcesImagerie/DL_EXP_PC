The ``as`` in the import statement allows to rename the module in the local context.

The module tifffile provides functions to read and write tiff-images. 

We also divide all intensity values in the input image by 255. This maps the intensity range from value between 0 and 255 to values between 0 and 1. Although the images are 16bit the maximum value they contain is 255.

The maping is in this case not strictly necessary, since all feature values (the intensities in a neighborhood of a pixel) already have the same range. However the optimizer in the training process works better on the range 0 to 1.

````python
import tifffile as tiff
from PIL import Image               # to create displayable images
from IPython.display import display # to display images
import numpy as np

inputPath = './training/spots.tif'
inputImage = tiff.imread(inputPath) / 255
im = Image.fromarray(np.uint8(inputImage*255))
display(im)
````

Tifffile opens the image as a numpy n-dimensional array. [numpy](https://numpy.org/) is an external library that provides n-dimensional arrays.
